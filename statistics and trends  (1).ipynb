# Data Distributions

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(12, 12))

# Create a function to plot histograms
def plot_hist(data, xlabel, title, subplot_index):
    plt.subplot(3, 3, subplot_index)
    sns.histplot(data, kde=True, color='skyblue')
    plt.xlabel(xlabel)
    plt.title(title)

# Plot histograms for each attribute
plot_hist(df['Size'], 'Size', 'Size Distribution', 1)
plot_hist(df['Weight'], 'Weight', 'Weight Distribution', 2)
plot_hist(df['Sweetness'], 'Sweetness', 'Sweetness Distribution', 3)
plot_hist(df['Crunchiness'], 'Crunchiness', 'Crunchiness Distribution', 4)
plot_hist(df['Juiciness'], 'Juiciness', 'Juiciness Distribution', 5)
plot_hist(df['Ripeness'], 'Ripeness', 'Ripeness Distribution', 6)
plot_hist(df['Acidity'], 'Acidity', 'Acidity Distribution', 7)

plt.tight_layout()
plt.show()
```

# Correlation Matrix

```python
# The correlation matrix is a table showing correlation coefficients between variables. 1 means positive correlation, -1 means negative correlation, 0 means no correlation
cols = df.columns[:-1]
correlation_matrix = df[cols].corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))  # mask to hide the upper triangle of the matrix
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', linewidths=1,
            linecolor='black')  # annot = True to show the values inside the cells


plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
